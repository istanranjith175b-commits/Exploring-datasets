```
import pandas as pd
df=pd.read_csv("employee_promotion.csv")
miss_count=df.isnull().sum()
cols_with_missing=miss_count[miss_count>0]
print("Colums with missing values and their counts:")
print(cols_with_missing)
miss_perce=(df.isnull().sum()/len(df))*100
print("\nPercentage of missing data per column:")
print(miss_perce)
highest_missing_col=miss_count.idxmax()
print(f"\nColumn with the highest missing entries: {highest_missing_col}")
nodup=df.duplicated().sum()
print(f"\nDuplicated records:{nodup}")
```
